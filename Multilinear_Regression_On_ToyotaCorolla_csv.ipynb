```python
# Renaming the columns name and making it short
data=cars.rename({'Age_08_04':'Age','cc':'CC','Quarterly_Tax':'QT'},axis=1)
data.head()
```

```python
raw_data=data[["Price","Age","KM","HP","CC","Doors","Gears","QT","Weight"]]
raw_data
```

```python
data = pd.get_dummies(raw_data,columns= ['Doors','HP','QT'])
data
```

```python
## Splitting Feature and Target Column as X and y
X = data.iloc[:,1:].values
y = data.iloc[:,0].values
```

```python
X
```

```python
y
```

```python
#Now we will spit our data into Train set and Test Set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state = 101)
```

```python
print(f"Shape of X_train = {X_train.shape}")
print(f"Shape of X_test = {X_test.shape}")
print(f"Shape of y_train = {y_train.shape}")
print(f"Shape of y_test = {y_test.shape}")
```

**Building Linear Regression Model**

```python
#Now we will build regression model on Training Set and Test it on our Test Set

from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X=X_train,y=y_train)
```

```python
#Now it's time to test the accuracy of the model on our Test Data
#this is very good accuracy on training set
lm.score(X_train,y_train)
```

```python
#Predicting the Test set results
Y_pred = lm.predict(X_test)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,Y_pred)
```

```python
score
```

**The R square value comes out to be 0.9116 which is a good result and hence we can say it is a good model.**
